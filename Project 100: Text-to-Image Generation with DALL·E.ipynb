```python
!pip install diffusers torch accelerate transformers
from diffusers import StableDiffusionPipeline
import torch

# Gunakan model Stable Diffusion
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")  # Gunakan GPU Colab

prompt = "A futuristic city skyline at sunset, with flying cars and neon lights"
image = pipe(prompt).images[0]

# Tampilkan gambar
image.show()

# Simpan
image.save("stable_diffusion_image.jpg")
print("Gambar disimpan!")
```
